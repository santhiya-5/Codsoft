```python
import numpy as numpy
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
df['species'].categories = pd.factorize(df['species'])
df.head()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
species_unique = df['species'].unique()
species_mapping = {species: i for i, species in enumerate(species_unique)}
species_numeric = df['species'].map(species_mapping)
ax.scatter(df.petal_length, df.petal_width,species_numeric)
ax.set_xlabel('petal_length in Cms')
ax.set_ylabel('petal_width in Cms')
ax.set_zlabel('species')
ax.set_zticks(range(len(species_unique)))
ax.set_zticklabels(species_unique)
plt.title('3D Scatter plot Example')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df.sepal_length, df.sepal_width,species_numeric)
ax.set_xlabel('sepal_length in Cms')
ax.set_ylabel('sepal_width in Cms')
ax.set_zlabel('species')
plt.title('3D Scatter plot Example')
plt.show()
```

```python
sns.scatterplot(data = df, x = 'sepal_length', y = 'sepal_width', hue = 'species')
```

```python
sns.scatterplot(data = df, x = 'petal_length', y = 'petal_width', hue = 'species')
```

```python
k_rng = range(1,10)
sse = []

for k in k_rng:
    km = KMeans(n_clusters = k)
    km.fit( df[['petal_length','petal_width']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('k_rng')
plt.ylabel('Sum of Squarred Errors')
plt.plot(k_rng,sse)
```

```python
km = KMeans(n_clusters = 3,random_state = 0)
y_predicted = km.fit_predict(df[['petal_length','petal_width']])
y_predicted
```

```python
df['cluster'] = y_predicted
df.head(5)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['species'] = le.fit_transform(df['species'])
df['cluster'] = le.fit_transform(df['cluster'])
cm = confusion_matrix(df['species'],df['cluster'])
cm
```

```python
 true_labels = df.species
 predicted_labels = df.cluster
 cm = confusion_matrix(true_labels, predicted_labels)
 class_labels = ['setosa', 'versicolor', 'virginica']

 plt.imshow(cm,interpolation='nearest',cmap=plt.cm.Blues)
 plt.title('Confusion Matrix')
 plt.colorbar()
 tick_marks = numpy.arange(len(class_labels))
 plt.xticks(tick_marks,class_labels,)
 plt.yticks(tick_marks,class_labels)
 plt.xlabel('Predicted Label')
 plt.ylabel('True Label')
 plt.show()



```

```python
x = df[['sepal_length','sepal_width','petal_length','petal_width']]
y = df['species']
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

```python
X_train,X_test,y_train,y_test
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled,X_test_scaled
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn
```

```python
knn.fit(X_train_scaled,y_train)
```

```python
y_pred = knn.predict(X_test_scaled)
species_names = df['species'].unique().astype(str).tolist()
report = classification_report(y_test,y_pred,target_names = species_names)
print(report)
```
